Question Generator


````python
question_prompt = f"""
You are an expert interview designer.

Based on these requirements:

{requirements_json}

Generate EXACTLY 6 interview questions.

Question Distribution:

1 Core Question
2 Technical Questions
2 Behavioural Questions
1 Situational Question

Return JSON ONLY.

Format:

[
    {{
        "question": "",
        "question_tag": "",
        "difficulty": "",
        "target_competency": [],
        "expected_scope": [],
        "follow_up_trigger": []
    }}
]

Rules:

CORE QUESTION
------------
Generate exactly 1.

Purpose:
- candidate introduction
- motivation
- career goals

Difficulty:
easy

TECHNICAL QUESTIONS
-------------------
Generate exactly 2.

Purpose:
- technical skills
- domain knowledge
- problem solving

Difficulty:
1 easy
1 medium

BEHAVIOURAL QUESTIONS
---------------------
Generate exactly 2.

Purpose:
- teamwork
- communication
- ownership
- conflict handling

Difficulty:
1 medium
1 hard

SITUATIONAL QUESTIONS
---------------------
Generate exactly 1.

Purpose:
- decision making
- critical thinking
- problem solving

Difficulty:
hard

Each question must contain:

question
question_tag
difficulty
target_competency
expected_scope
follow_up_trigger

Allowed Tags:

core
technical
behavioural
situational

Return VALID JSON ONLY.
No markdown.
No explanation.
"""
````

````python
response = ollama.chat(
    model="qwen3:8b",
    messages=[
        {
            "role":"user",
            "content":question_prompt
        }
    ]
)

questions_result = response["message"]["content"]

print(questions_result)
````

job question json save

````python
import json
import re

clean_output = re.sub(
    r"```json|```",
    "",
    questions_result
).strip()

questions_json = json.loads(clean_output)

print("Questions loaded successfully")
print(f"Total Questions: {len(questions_json)}")
````

EXPECTED ANSWER GENERATOR

````python
expected_answer_prompt = f"""
You are an expert AI Interview Designer.

Based on these interview questions:

{questions_json}

Generate a JSON array.

For EACH question generate:

[
  {{
    "question": "",
    "question_tag": "",
    "difficulty": "",
    "competency": [],

    "weight": 0,

    "score_breakdown": {{
        "keywords": 30,
        "scope_coverage": 30,
        "relevance": 20,
        "communication": 20
    }},

    "ideal_answer": "",

    "keywords": [],

    "expected_scope": [],

    "follow_up_trigger": [],

    "scoring_criteria": {{
        "poor": "",
        "average": "",
        "excellent": ""
    }}
  }}
]

Rules:

1. Preserve:
   - question
   - question_tag
   - difficulty

2. competency:
   Identify competencies being assessed.

3. Weight:
   easy = 5
   medium = 10
   hard = 15

4. ideal_answer:
   realistic interview answer
   80-150 words

5. keywords:
   generate 5-8 important keywords

6. expected_scope:
   generate 3-6 expected discussion points

7. follow_up_trigger:
   examples:
   [
      "missing_example",
      "unable_to_explain",
      "weak_reasoning",
      "incomplete_answer"
   ]

8. scoring_criteria:
   poor
   average
   excellent

9. score_breakdown must always be:

{{
    "keywords": 30,
    "scope_coverage": 30,
    "relevance": 20,
    "communication": 20
}}

10. Return VALID JSON ONLY

No markdown.
No explanation.
"""
````

````python
response = ollama.chat(
    model="qwen3:8b",
    messages=[
        {
            "role": "user",
            "content": expected_answer_prompt
        }
    ]
)

expected_answers_result = response["message"]["content"]
````

````python
import json
import re

cleaned = re.sub(
    r"```json|```",
    "",
    expected_answers_result
).strip()

expected_answers_json = json.loads(cleaned)

print(
    f"Generated {len(expected_answers_json)} expected answers."
)
````

expected answer json save

````python
with open(
    "expected_answers.json",
    "w",
    encoding="utf-8"
) as f:

    json.dump(
        expected_answers_json,
        f,
        indent=4,
        ensure_ascii=False
    )

print("expected_answers.json saved")
````

````python
for i, item in enumerate(expected_answers_json, start=1):

    print("="*80)

    print(f"Question {i}")

    print()
    print("Question:")
    print(item["question"])

    print()
    print("Keywords:")
    print(item["keywords"])

    print()
    print("Expected Scope:")
    print(item["expected_scope"])

    print()
    print("Excellent Answer:")
    print(item["scoring_criteria"]["excellent"])

    print()
````

````python
for q in expected_answers_json:

    print(
        f"{q['difficulty']} -> Weight {q['weight']}"
    )
````

````python
from collections import Counter

competencies = []

for q in expected_answers_json:
    competencies.extend(
        q["competency"]
    )

counter = Counter(competencies)

print("\nCOMPETENCY DISTRIBUTION\n")

for comp, count in counter.items():
    print(f"{comp}: {count}")
````
